```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import matplotlib.patches as mpatches
import math
from matplotlib.ticker import MaxNLocator
import matplotlib.ticker as mtick
import matplotlib as mpl


plt.rcParams.update({'font.size': 20})
%config InlineBackend.figure_format = 'retina'
```

```
f = 22.5 # mL/(hr*collector)
N = 88   # Total fractions
v = 1.   # volume per fraction
k = np.arange(1, 20) # number of collectors

hours = N*v / (k*f)
cost_per = 67.02
cost = [i*cost_per for i in range(1, 20)]
```

```
fig, ax = plt.subplots(figsize=(7, 7))
ax2 = ax.twinx()

ax.scatter(k, hours, c='k', s=100, label="Time")

ax.set_xlabel("Number of fraction collectors")
ax.set_ylabel("Experiment time [hr]")
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.xaxis.set_major_locator(MaxNLocator(integer=True))


ax2.scatter(k, cost, c='grey', s=100, label="Cost")

ax2.set_ylabel("Estimated cost [$]")
# ax2.yaxis.set_major_locator(MaxNLocator(integer=True))
ax2.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

grey_patch = mpatches.Patch(color='grey', label='Cost')
black_patch = mpatches.Patch(color='black', label='Time')
ax.legend(handles=[black_patch, grey_patch], loc="upper left", bbox_to_anchor=(0.1, 1))

fig.savefig("time_cost_scale.png", bbox_inches='tight', dpi=100)


fig.show()
```
